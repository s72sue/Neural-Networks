```python
theta_1 = 3.0
theta_2 = 2.0
def bcm_rule_1(x):
    return x * (x - theta_1)

def bcm_rule_2(x):
    if x > theta_2:
        return np.tanh([(x-theta_2)*2.0])[0]
    return x * (x - theta_2)

x = np.arange(0.0,5.0,0.1)
y1 = x * (x - theta_1)
y2 = x * (x - theta_2)
y2[x > theta_2] = np.tanh([(x[x > theta_2]-theta_2)*2.0])[0]

plt.figure(1, figsize=(9, 4.5))
plt.subplot(121)
plt.plot(x, y1, color='k', linewidth=2)
plt.axhline(linestyle='--', linewidth=1, color='k')
plt.title(r"$y = x (x - \theta)$", fontsize=14)
plt.axis((0.0, 5.0, -2.5, 6.0))
plt.xlabel("$a_j$", fontsize='large')
plt.ylabel("$\phi(a_j, </span><span class="se">\\</span><span class="s">theta)$", fontsize='large')
plt.yticks([0])
plt.xticks([theta_1], [r"$\theta$"])

plt.subplot(122)
plt.plot(x, y2, color='k', linewidth=2)
plt.axhline(linestyle='--', linewidth=1, color='k')
plt.title(r"if $x > \theta,$ $y = \tanh (x)$", fontsize=14)
plt.xlabel("$a_j$", fontsize='large')
plt.axis((0.0, 5.0, -1.5, 1.5))
plt.yticks([0])
plt.xticks([theta_2], [r"$\theta$"]);
```

```python
import nengo
import numpy as np

model = nengo.Network()
with model:
    sin = nengo.Node(lambda t: np.sin(t*4))
    
    pre = nengo.Ensemble(100, dimensions=1)
    post = nengo.Ensemble(100, dimensions=1)

    nengo.Connection(sin, pre)
    conn = nengo.Connection(pre, post, solver=nengo.solvers.LstsqL2(weights=True))
    conn.learning_rule_type = nengo.BCM()
    conn.learning_rule_type.learning_rate *= 0.1
```

```python
from nengo_gui.ipython import IPythonViz
IPythonViz(model, cfg="BCM.ipynb.cfg")
```

```python
with model:
    trans_p = nengo.Probe(conn, 'transform', synapse=None, sample_every=1.0)
sim = nengo.Simulator(model)
sim.run(100.0)
```
