```python
import numpy as np
x = [1,3,4,7]
w = [1,3,4,7]
N = 4
threshold = 6
net = 0;
for i in range(N):
    net = net + (x[i] * w[i])  #Calculates weighted Sum

    
#Finds the output based on the net
if(net >= threshold):
    #output =  1    
    output = 1 / (1 + np.exp(-net));
if(net <  threshold):
    #output = -1
    output = -1 / (1 + np.exp(-net));
    
print output    
```

```python
def main():
    inputs = generateData(5)
    print inputs

def generateData(n):
    """ 
    generates a 2D linearly separable dataset with n samples. 
    The third element of the sample is the label
    """
    xb = (np.random.rand(n)*2-1)/2-0.5
    yb = (np.random.rand(n)*2-1)/2+0.5
    xr = (np.random.rand(n)*2-1)/2+0.5
    yr = (np.random.rand(n)*2-1)/2-0.5
    inputs = []
    for i in range(len(xb)):
        inputs.append([xb[i],yb[i],1])
        inputs.append([xr[i],yr[i],-1])
    return inputs

if __name__ == "__main__":
    main()
```

```python
%pylab inline
import nengo
from nengo.dists import Uniform
from nengo.utils.ensemble import response_curves
import numpy as np

N = 10 #number of neurons
D = 1
model = nengo.Network('Perceptron')
weights = np.random.rand(N)
#start = [1,1]   #np.random.rand(N)

with model:
    
    def stim(t):
        if t<=1 or 2.5<t<=3.5:
            return 0.5
        elif 1<t<=2 or 3.5<t<=4.5:
            return -0.5
        else:
            return 0
        
    def stim_ans(t):
        if t<=1:
            return 1
        elif 1<t<=2:
            return -1
        else:
            return 0    
    
    
    stim = nengo.Node(output= stim, size_out=D)
   
    stim_ans = nengo.Node(output=stim_ans, size_out=1)
    
    
    input = nengo.Ensemble(N, dimensions=D) #,  
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    
    answer = nengo.Ensemble(N, dimensions=1) #, 
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    
    error = nengo.Ensemble(N, dimensions=1) #, 
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    correct_answer = nengo.Ensemble(N, dimensions=1) #, 
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))

    
    nengo.Connection(stim, input, synapse=0) 
    nengo.Connection(stim_ans, correct_answer, synapse=0)
    
    nengo.Connection(answer, error, transform=-1)
    nengo.Connection(correct_answer, error)
    error_conn = nengo.Connection(error, answer, modulatory=True)
    conn = nengo.Connection(input, answer, function=lambda x: np.random.random(1), solver=nengo.solvers.LstsqL2(weights=True)) 
    conn.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-4)
    
    def inhibit(t):
        return 2.0 if t > 2.0 else 0.0

    inhib = nengo.Node(inhibit)
    nengo.Connection(inhib, error.neurons, transform=[[-1]] * error.n_neurons)
    
```

```python
with model:
    input_p = nengo.Probe(input, synapse=0.1)
    answer_p = nengo.Probe(answer, synapse=0.1)
    correct_answer_p = nengo.Probe(correct_answer, synapse=0.1)
    error_p = nengo.Probe(error, synapse=0.1)
```

```python
sim = nengo.Simulator(model)
sim.run(5)
t = sim.trange()
```

```python
figure()
plot(*response_curves(answer, sim))
title("Response curves")

figure()
plot(t, sim.data[input_p])
title("Input (Stim)")

figure()
plot(t, sim.data[answer_p])
title("Answer")

figure()
plot(t, sim.data[correct_answer_p])
title("Correct Answer")

figure()
plot(t, sim.data[error_p])
title("Error")
```
