```python
%pylab inline
import nengo
from nengo.dists import Uniform
from nengo.utils.ensemble import response_curves, tuning_curves
import numpy as np
import matplotlib.pyplot as plt

    
def generateData(n, function):
    """ 
    generates a 2D linearly separable bianry dataset 
    """
    inputs = []
    
    if function == "XOR":
        for i in range(n):
            inputs.append([0,0,1,[0]])
            inputs.append([0,1,1,[1]])
            inputs.append([1,0,1,[1]])
            inputs.append([1,1,1,[0]])    
    return inputs


samples = 40
sets = samples / 4
data = generateData(sets, "XOR")

for x in data:
    if x[3][0] == 1:
        plt.plot(x[0], x[1], 'or')
    else:
        plt.plot(x[0], x[1], 'ob') 
```

```python
N = 30 #number of neurons
D_inp = 3
D_out = 1
D_hl = 3
factor = 3.0/4.0 #11.0/12.0 
sim_time = samples
neuron_mode = nengo.Sigmoid(tau_ref=0.002) #nengo.LIF()
model = nengo.Network('XOR Network')

weights = np.random.rand(D_inp,D_out).transpose()    

with model:  
    
    def stim(t):
        for i in range(samples):
            if int(t) % samples <= i:
                #print [data[i][0], data[i][1], data[i][2]], "time:", t
                return [data[i][0], data[i][1], data[i][2]]
        return 0
    
        
    def stim_ans(t):
        for i in range(samples):
            if int(t) % samples <= i:
                #print data[i][3], "time:",  t
                return data[i][3]
        return 0   
    
     
    stim = nengo.Node(output=stim, size_out=D_inp)
    stim_ans = nengo.Node(output=stim_ans, size_out=D_out) 
    
    
    input = nengo.Ensemble(N*D_inp, dimensions=D_inp, neuron_type=neuron_mode) 
    answer = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)
    error = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)  
    correct_answer = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)
    
    nengo.Connection(stim, input) 
    nengo.Connection(stim_ans, correct_answer)
    
    nengo.Connection(answer, error, transform=-1)
    nengo.Connection(correct_answer, error)

    conn = nengo.Connection(input, answer, transform=weights, solver=nengo.solvers.LstsqL2(weights=True))
    nengo.Connection(answer, error, transform=-1)
    nengo.Connection(correct_answer, error)
    error_conn = nengo.Connection(error, answer, modulatory=True)
    
    conn.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-6)
    
    def inhibit(t):
        return 2.0 if t > int(factor*samples) else 0.0
     
    inhib = nengo.Node(inhibit)
    nengo.Connection(inhib, error.neurons, transform=[[-1]] * error.n_neurons)
    
    actual_error = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)  
    
    #Actual Error
    nengo.Connection(answer ,actual_error, transform=-1)
    nengo.Connection(correct_answer, actual_error)
       
```

```python
with model:
    input_p = nengo.Probe(input, synapse=0.1)
    answer_p = nengo.Probe(answer, synapse=0.1)
    correct_answer_p = nengo.Probe(correct_answer, synapse=0.1)
    error_p = nengo.Probe(error, synapse=0.1)
    #error_p = nengo.Probe(error, synapse=0.1)
    #output_p = nengo.Probe(output, synapse=0.1)
    
    #conn_p = nengo.Probe(conn, synapse=0.1, attr='transform', sample_every=30)
    
    actual_error = nengo.Probe(actual_error, synapse=0.1) 
```

```python
sim = nengo.Simulator(model)
sim.run(sim_time)
t = sim.trange()
```

```python
plt.figure()
plt.plot(*response_curves(answer, sim))
plt.title("Response curves")
plt.show()

fig = figure(figsize=(18,18))
p1 = fig.add_subplot(4,1,1)
p1.plot(t, sim.data[answer_p])
p1.set_title("Answer")
#p1.set_xlim(int(factor*sim_time),sim_time)
#p1.set_xlim(2900,sim_time)

p2 = fig.add_subplot(4,1,2)
p2.plot(t, sim.data[correct_answer_p])
p2.set_title("Correct Answer")
#p2.set_xlim(int(factor*sim_time),sim_time)
#p2.set_xlim(2900,sim_time)

p3 = fig.add_subplot(4,1,3)
p3.plot(t, sim.data[error_p])
p3.set_title("Error_1")

p4 = fig.add_subplot(4,1,4)
p4.plot(t, sim.data[actual_error])
p4.set_title("Acutal Error")

figure()
plot(t, sim.data[input_p])
title("Input")
xlim(0,20)
```
