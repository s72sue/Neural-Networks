Now we will use three different ensembles in the input layer to see whether it still solves XOR in two layers. 

Conclusion: It was not successful at solving an XOR in two layers. A hidden layer is needed.

```python
%pylab inline
import nengo
from nengo.dists import Uniform
from nengo.utils.ensemble import response_curves, tuning_curves
import numpy as np
import matplotlib.pyplot as plt

    
def generateData(n, function):
    """ 
    generates a 2D linearly separable bianry dataset 
    """
    inputs = []
    
    if function == "XOR":
        for i in range(n):
            inputs.append([0,0,1,[0]])
            inputs.append([0,1,1,[1]])
            inputs.append([1,0,1,[1]])
            inputs.append([1,1,1,[0]])    
    return inputs


samples = 240 #140
sets = samples / 4
data = generateData(sets, "XOR")

for x in data:
    if x[3][0] == 1:
        plt.plot(x[0], x[1], 'or')
    else:
        plt.plot(x[0], x[1], 'ob') 
```

```python
N = 30 #number of neurons
D_inp = 1
D_out = 1
D_hl = 3
factor = 3.0/4.0 #11.0/12.0 
sim_time = samples
neuron_mode = nengo.Sigmoid(tau_ref=0.002) #nengo.LIF()
model = nengo.Network('XOR Network')

weights = np.random.rand(D_inp,D_out).transpose()    

with model:  
    
    def stim1(t):
        for i in range(samples):
            if int(t) % samples <= i:
                #print [data[i][0]], "time:", t
                return data[i][0]
        return 0
    
    def stim2(t):
        for i in range(samples):
            if int(t) % samples <= i:
                #print [data[i][1]], "time:", t
                return data[i][1]
        return 0
    
    def stim3(t):
        for i in range(samples):
            if int(t) % samples <= i:
                #print [data[i][2]], "time:", t
                return data[i][2]
        return 0
    
        
    def stim_ans(t):
        for i in range(samples):
            if int(t) % samples <= i:
                #print data[i][3], "time:",  t
                return data[i][3]
        return 0   
    
    #Input layer ensembles
    input1 = nengo.Ensemble(N*D_inp, dimensions=D_inp, neuron_type=neuron_mode) 
    input2 = nengo.Ensemble(N*D_inp, dimensions=D_inp, neuron_type=neuron_mode) 
    input3 = nengo.Ensemble(N*D_inp, dimensions=D_inp, neuron_type=neuron_mode) 
    
    #Provide input to the model
    stim_ans = nengo.Node(output=stim_ans, size_out=D_out) 
    stim1 = nengo.Node(output=stim1, size_out=D_inp)
    stim2 = nengo.Node(output=stim2, size_out=D_inp)
    stim3 = nengo.Node(output=stim3, size_out=D_inp)
    nengo.Connection(stim1, input1) 
    nengo.Connection(stim2, input2)
    nengo.Connection(stim3, input3)
    
    # Error based on which learning will occur for all three connections
    answer = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)
    error = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)  
    correct_answer = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)
    
    nengo.Connection(stim_ans, correct_answer)
    nengo.Connection(answer, error, transform=-1)
    nengo.Connection(correct_answer, error)

    conn1 = nengo.Connection(input1, answer, transform=weights, solver=nengo.solvers.LstsqL2(weights=True))
    conn2 = nengo.Connection(input2, answer, transform=weights, solver=nengo.solvers.LstsqL2(weights=True))
    conn3 = nengo.Connection(input3, answer, transform=weights, solver=nengo.solvers.LstsqL2(weights=True))
    error_conn = nengo.Connection(error, answer, modulatory=True)
    
    conn1.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-6)
    conn2.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-6)
    conn3.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-6)
    
    def inhibit(t):
        return 2.0 if t > int(factor*samples) else 0.0
     
    inhib = nengo.Node(inhibit)
    nengo.Connection(inhib, error.neurons, transform=[[-1]] * error.n_neurons)
    
    #Actual Error
    actual_error = nengo.Ensemble(N*D_out, dimensions=D_out, neuron_type=neuron_mode)  
    nengo.Connection(answer ,actual_error, transform=-1)
    nengo.Connection(correct_answer, actual_error)
       
```

```python
with model:
    answer_p = nengo.Probe(answer, synapse=0.1)
    correct_answer_p = nengo.Probe(correct_answer, synapse=0.1)
    error_p = nengo.Probe(error, synapse=0.1) 
    actual_error = nengo.Probe(actual_error, synapse=0.1) 
```

```python
sim = nengo.Simulator(model)
sim.run(sim_time)
t = sim.trange()
```
