```python
%pylab inline
import nengo
from nengo.dists import Uniform
from nengo.utils.ensemble import response_curves
import numpy as np


# ---------------------------------------------------------------------------------------------------------------------

def generateData(n):
    """ 
    generates a 2D linearly separable dataset with n samples. 
    The third element of the sample is the label
    """
    xb = (np.random.rand(n)*2-1)/2-0.5
    yb = (np.random.rand(n)*2-1)/2+0.5
    xr = (np.random.rand(n)*2-1)/2+0.5
    yr = (np.random.rand(n)*2-1)/2-0.5
    inputs = []
    for i in range(len(xb)):
        inputs.append([xb[i],yb[i],2])
        inputs.append([xr[i],yr[i],-2])
    return inputs


# ---------------------------------------------------------------------------------------------------------------------

N = 20 #number of neurons
D = 1
samples = 250
model = nengo.Network('Perceptron')
weights1 = np.random.rand(D,D)
weights2 = np.random.rand(D,D)
data = generateData(samples)

with model:  
    """def stim1(t):
        if t<=1 or 2<t<=3:
            return -0.50320063987540053
        elif 1<t<=2 or 3<t<=4:
            return 0.83865901369051798
        else:
            return 0
        
    def stim2(t):
        if t<=1 or 2<t<=3:
            return 0.24527793152044131
        elif 1<t<=2 or 3<t<=4:
            return -0.1059601755781634
        else:
            return 0    """
    
    def stim1(t):
        for i in range(samples):
            if t < i:
                return data[i][0]
        return 0
        
    def stim2(t):
        for i in range(samples):
            if t < i:
                return data[i][1]
        return 0
    
        
    def stim_ans(t):
        for i in range(samples):
            if t < i:
                return data[i][2]
        return 0  
    
    
    stim1 = nengo.Node(output= stim1, size_out=D)
    stim2 = nengo.Node(output= stim2, size_out=D)
    stim_ans = nengo.Node(output=stim_ans, size_out=D)
    
    
    input1 = nengo.Ensemble(N, dimensions=D) #,  
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
                
    input2 = nengo.Ensemble(N, dimensions=D) #,  
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))            
    
    
    answer = nengo.Ensemble(N, dimensions=D, radius=2) #, 
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    
    error = nengo.Ensemble(N, dimensions=D, radius=2) #, 
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    correct_answer = nengo.Ensemble(N, dimensions=D, radius=2) #, 
                         #max_rates=Uniform(.999,.999),
                         #intercepts=Uniform(-1,-1),
                         #neuron_type=nengo.neurons.Sigmoid(tau_ref=1))

    
    nengo.Connection(stim1, input1) 
    nengo.Connection(stim2, input2) 
    nengo.Connection(stim_ans, correct_answer)
    
    nengo.Connection(answer, error, transform=-1)
    nengo.Connection(correct_answer, error)
    error_conn = nengo.Connection(error, answer, modulatory=True)
    
    conn1 = nengo.Connection(input1, answer, transform=weights1, solver=nengo.solvers.LstsqL2(weights=True)) 
    conn2 = nengo.Connection(input2, answer, transform=weights2, solver=nengo.solvers.LstsqL2(weights=True))
    #conn = nengo.Connection(input, answer, function=lambda x: np.random.random(1), solver=nengo.solvers.LstsqL2(weights=True)) 
    
    conn1.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-6)
    conn2.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-6)
    
    def inhibit(t):
        return 2.0 if t > 200.0 else 0.0

    inhib = nengo.Node(inhibit)
    nengo.Connection(inhib, error.neurons, transform=[[-1]] * error.n_neurons)
    
```

```python
with model:
    input1_p = nengo.Probe(input1, synapse=0.1)
    input2_p = nengo.Probe(input2, synapse=0.1)
    answer_p = nengo.Probe(answer, synapse=0.1)
    correct_answer_p = nengo.Probe(correct_answer, synapse=0.1)
    error_p = nengo.Probe(error, synapse=0.1)
```

```python
sim = nengo.Simulator(model)
sim.run(250)
t = sim.trange()
```

```python
figure()
plot(*response_curves(answer, sim))
title("Response curves")

fig = figure(figsize=(18,10))
p1 = fig.add_subplot(3,1,1)
p1.plot(t, sim.data[answer_p])
p1.set_title("Answer")

figure()
plot(t, sim.data[correct_answer_p])
title("Correct Answer")

figure()
plot(t, sim.data[error_p])
title("Error")
```
