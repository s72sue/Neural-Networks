```python
%pylab inline
import nengo
from nengo.dists import Uniform
from nengo.utils.ensemble import response_curves
import numpy as np
import matplotlib.pyplot as plt


def generateData(n):
    """ 
    generates a 2D linearly separable dataset with n samples. 
    The third element of the sample is the label
    """
    xg = (np.random.rand(n)*2-1)/2-0.5
    yg = (np.random.rand(n)*2-1)/2+0.5
    xb = (np.random.rand(n)*2-1)/2+0.5
    yb = (np.random.rand(n)*2-1)/2+0.5
    xr = (np.random.rand(n)*2-1)/2-0.5
    yr = (np.random.rand(n)*2-1)/2-0.5
    
    inputs = []
    for i in range(len(xb)):
        inputs.append([xb[i],yb[i],[0,0,1]])
        inputs.append([xr[i],yr[i],[1,0,0]])
        inputs.append([xg[i],yg[i],[0,1,0]])
    return inputs


```

```python
N = 100 #number of neurons
D_inp = 3
D_hl = 4
D_out = 3
samples = 50
model = nengo.Network('Backpropagation')
data = generateData(samples)

for x in data:
    if x[2][0] == 1:
        plt.plot(x[0], x[1], 'or')
    elif x[2][1] == 1:
        plt.plot(x[0], x[1], 'og')
    else:
        plt.plot(x[0], x[1], 'ob') 
        

weights_2 = np.random.randn(D_inp,D_hl).transpose()    
weights_1 = np.random.randn(D_hl,D_out).transpose()  

print "Conn1 weights: ", weights_1
print "Conn2 weights: ", weights_2


with model:  
    def stim(t):
        for i in range(samples):
            if t < i:
                return [data[i][0], data[i][1], 1]
        return 0
    
        
    def stim_ans(t):
        for i in range(samples):
            if t < i:
                return data[i][2]
        return 0  
    
    def compute(x):
        return x * (1-x)
    
    
    stim = nengo.Node(output=stim, size_out=D_inp)
    stim_ans = nengo.Node(output=stim_ans, size_out=D_out)
    
    
    """input = nengo.Ensemble(N, dimensions=D_inp) 
    hiddenlayer = nengo.Ensemble(N, dimensions=D_hl)
    answer = nengo.Ensemble(N, dimensions=D_out)
    error_1 = nengo.Ensemble(N, dimensions=D_out, radius=2)  
    correct_answer = nengo.Ensemble(N, dimensions=D_out, radius=2)"""
    
    input = nengo.Ensemble(N, dimensions=D_inp,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    hiddenlayer = nengo.Ensemble(N, dimensions=D_hl,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    answer = nengo.Ensemble(N, dimensions=D_out,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    error_1 = nengo.Ensemble(N, dimensions=D_out,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    correct_answer = nengo.Ensemble(N, dimensions=D_out,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    
    nengo.Connection(stim, input) 
    nengo.Connection(stim_ans, correct_answer)
    
    conn_2 = nengo.Connection(input, hiddenlayer, transform=weights_2, solver=nengo.solvers.LstsqL2(weights=True))
    conn_1 = nengo.Connection(hiddenlayer, answer, transform=weights_1, solver=nengo.solvers.LstsqL2(weights=True)) 
    
    nengo.Connection(answer, error_1, transform=-1)
    nengo.Connection(correct_answer, error_1)
    error_conn1 = nengo.Connection(error_1, answer, modulatory=True)
    
    conn_1.learning_rule_type = nengo.PES(error_conn1, learning_rate=1e-6)
    
    
    # Error for the hidden layer
    #error_2 = nengo.Ensemble(N, dimensions=D_hl, radius=2)  
    error_2 = nengo.Ensemble(N, dimensions=D_hl,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    #wt . delta of next layer
    #inter1 = nengo.Ensemble(N, dimensions=D_hl, radius=2)  
    inter1 = nengo.Ensemble(N, dimensions=D_hl,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    nengo.Connection(error_1, inter1, transform=conn_1.transform.transpose(), solver=nengo.solvers.LstsqL2(weights=True))
    
    # contains a * 1-a = [ hiddenlayer activity * (1-hidden layer activity) ]
    #inter2 = nengo.Ensemble(N, dimensions=D_hl, radius=2)
    inter2 = nengo.Ensemble(N, dimensions=D_hl,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    nengo.Connection(hiddenlayer, inter2, function=compute)
    
    #inter3 = nengo.Ensemble(N, dimensions=2*D_hl, radius=2)
    inter3 = nengo.Ensemble(N, dimensions=2*D_hl,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    #product = nengo.Ensemble(N, dimensions=D_hl, radius=2)
    product = nengo.Ensemble(N, dimensions=D_hl,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    
    A = np.zeros((D_hl,2*D_hl))
    A[:,0:D_hl] = 1
    
    B = np.zeros((D_hl,2*D_hl))
    B[:,D_hl:2*D_hl] = 1
    
    nengo.Connection(inter1, inter3, transform=(A/D_hl).transpose())
    nengo.Connection(inter2, inter3, transform=(B/D_hl).transpose())
    nengo.Connection(inter3, product, function=lambda x: x[0:D_hl]*x[D_hl:2*D_hl])
    nengo.Connection(product, error_2)
    
    error_conn2 = nengo.Connection(error_2, hiddenlayer, modulatory=True)
    conn_2.learning_rule_type = nengo.PES(error_conn2, learning_rate=1e-6)
       
    def inhibit(t):
        return 2.0 if t > 30.0 else 0.0
     
    inhib = nengo.Node(inhibit)
    nengo.Connection(inhib, error_1.neurons, transform=[[-1]] * error_1.n_neurons)
    nengo.Connection(inhib, error_2.neurons, transform=[[-1]] * error_2.n_neurons)
```

```python
with model:
    input1_p = nengo.Probe(input, synapse=0.1)
    answer_p = nengo.Probe(answer, synapse=0.1)
    correct_answer_p = nengo.Probe(correct_answer, synapse=0.1)
    error1_p = nengo.Probe(error_1, synapse=0.1)
    error2_p = nengo.Probe(error_2, synapse=0.1)
    #output_p = nengo.Probe(output, synapse=0.1)
    
    #conn_p = nengo.Probe(conn, synapse=0.1, attr='transform', sample_every=30)
```

```python
sim = nengo.Simulator(model)
sim.run(50.0)
t = sim.trange()
```
