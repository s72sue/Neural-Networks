```python
%pylab inline
import nengo
from nengo.dists import Uniform
from nengo.utils.ensemble import response_curves
import numpy as np

# ---------------------------------------------------------------------------------------------------------------------
# Randomly generated data sets with bias

class data:  
    def __init__(self, N):
        # Random linearly separated data
        xA,yA,xB,yB = [np.random.uniform(-1, 1) for i in range(4)]
        self.V = np.array([xB*yA-xA*yB, yB-yA, xA-xB])
        self.X = self.generate_points(N)
 
    def generate_points(self, N):
        X = []
        for i in range(N):
            x1,x2 = [np.random.uniform(-1, 1) for i in range(2)]
            x = np.array([1,x1,x2])
            s = int(np.sign(self.V.T.dot(x)))
            p = [x[1],x[2],s]
            X.append(p)
        return X


# ---------------------------------------------------------------------------------------------------------------------
# Fixed data set without bias

def generateData(n):
    """ 
    generates a 2D linearly separable dataset with n samples. 
    The third element of the sample is the label
    """
    choice = numpy.random.random_integers(1,2)
    
    if choice == 1:
        xb = (np.random.rand(n)*2-1)/2-0.5
        yb = (np.random.rand(n)*2-1)/2+0.5
        xr = (np.random.rand(n)*2-1)/2+0.5
        yr = (np.random.rand(n)*2-1)/2-0.5
    else:
        xb = (np.random.rand(n)*2-1)/2+0.5
        yb = (np.random.rand(n)*2-1)/2+0.5
        xr = (np.random.rand(n)*2-1)/2-0.5
        yr = (np.random.rand(n)*2-1)/2-0.5
    
    inputs = []
    for i in range(len(xb)):
        inputs.append([xb[i],yb[i],1])
        inputs.append([xr[i],yr[i],-1])
    return inputs


# ---------------------------------------------------------------------------------------------------------------------
# Reference weights [-1,1]: http://web.fhnw.ch/plattformen/ml/vorlesungsunterlagen/networking/mlpi.pdf


N = 100    #number of neurons
D = 3
samples = 50
model = nengo.Network('Perceptron')

#data = generateData(samples)
data = data(samples).X

for x in data:
    if x[2] == 1:
        plt.plot(x[0], x[1], 'ob')
    else:
        plt.plot(x[0], x[1], 'or') 
```

```python
weights = np.random.randn(1,D)
#weights = np.random.uniform(-0.77, 0.77, (1,D))
weights = np.array([[0.000001, -0.000001, 0.00002]])
print "wegiths:", weights


with model:  
    def stim(t):
        for i in range(samples):
            if t < i:
                return [data[i][0], data[i][1], 1]
        return 0
    
        
    def stim_ans(t):
        for i in range(samples):
            if t < i:
                return data[i][2]
        return 0  
    
    
    stim = nengo.Node(output=stim, size_out=D)
    stim_ans = nengo.Node(output=stim_ans, size_out=1)
    
    
    input = nengo.Ensemble(N, dimensions=D)                
    answer = nengo.Ensemble(N, dimensions=1, radius=2) 
    error = nengo.Ensemble(N, dimensions=1, radius=2)  
    correct_answer = nengo.Ensemble(N, dimensions=1, radius=2) 
    
    
    """input = nengo.Ensemble(N, dimensions=D,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    answer = nengo.Ensemble(N, dimensions=1,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    error = nengo.Ensemble(N, dimensions=1,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))
    
    correct_answer = nengo.Ensemble(N, dimensions=1,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))"""
    
    nengo.Connection(stim, input) 
    nengo.Connection(stim_ans, correct_answer)
    
    nengo.Connection(answer, error, transform=-1)
    nengo.Connection(correct_answer, error)
    error_conn = nengo.Connection(error, answer, modulatory=True)
    
    conn = nengo.Connection(input, answer, transform=weights, solver=nengo.solvers.LstsqL2(weights=True))  
    conn.learning_rule_type = nengo.PES(error_conn, learning_rate=1e-12)
    
    def inhibit(t):
        return 2.0 if t > 30.0 else 0.0

    inhib = nengo.Node(inhibit)
    nengo.Connection(inhib, error.neurons, transform=[[-1]] * error.n_neurons)
    
    output = nengo.Ensemble(N, dimensions=1,  
                     max_rates=Uniform(.999,.999),
                     intercepts=Uniform(-1,-1),
                     neuron_type=nengo.neurons.Sigmoid(tau_ref=1))

    
    nengo.Connection(answer, output)
```

```python
with model:
    input1_p = nengo.Probe(input, synapse=0.1)
    answer_p = nengo.Probe(answer, synapse=0.1)
    correct_answer_p = nengo.Probe(correct_answer, synapse=0.1)
    error_p = nengo.Probe(error, synapse=0.1)
    output_p = nengo.Probe(output, synapse=0.1)
    
    conn_p = nengo.Probe(conn, synapse=0.1, attr='transform', sample_every=50)
```

```python
sim = nengo.Simulator(model)
sim.run(50.0)
t = sim.trange()
```
